```python
data_div = driver.find_element_by_id('datagrid')
data_html = data_div.get_attribute('innerHTML')
```

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(data_html)
```

```python
import pandas as pd
```

```python
head = [t.text.replace("▼"," ").replace("▲"," ") for t in soup.thead.find_all("th")]
```

```python
head
```

```python
len(head)
```

```python
dataframe = pd.DataFrame(columns=head)
```

```python
dataframe.drop("", axis=1)
```

```python
dataframe.drop(" ", axis=1)
```

```python
context = []
i=[]
for t in soup.tbody.find_all("tr"):
    for a in t.find_all("td"):
        context.append(a.text.replace("\xa0",""))
```

```python
context_b = []
for i in range(49):
    a= context[i*(34):(i+1)*34]
    context_b.append(a)
```

```python
for i in range(49):
    dataframe.loc[i]=context_b[i]
dataframe
```
