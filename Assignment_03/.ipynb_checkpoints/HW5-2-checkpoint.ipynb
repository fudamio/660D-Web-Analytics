```python
import requests
from bs4 import BeautifulSoup

FREELANCE_COM = 'https://www.freelancer.com'  # main Freelance Page
ALL_USERS = '/freelancers/skills/all/{}'      # all users (based on page number)

PRETEND_AGENT = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.7; rv:11.0) Gecko/20100101 Firefox/11.0'
```

```python
df1=[]  #number of recommendations
df2=[]  #earnings
for j in range(50):
    url = FREELANCE_COM + ALL_USERS.format(j)
    content = requests.get(url, headers={'User-Agent': PRETEND_AGENT}).content
    soup = BeautifulSoup(content, "lxml")
    all_links = [x.get('href') for x in soup.find_all('a')]
    user_links = [x for x in all_links if x and x.find('/u/') >= 0]
    
    for i in user_links:
        user = BeautifulSoup(requests.get(FREELANCE_COM + i).content, "lxml")
        df1.append(user.find('span', {'data-target': "recommend-btn-val"}).text)
        df2.append(user.find('span', {'class': 'Earnings-label'}).text.strip())
```

```python
import numpy as np
import pandas as pd
recommendations = np.array(df1)
earnings = np.array(df2)
```

```python
#remove duplicates for recommendations
df11 = [item for index, item in enumerate(df1) if index % 3 != 0]
recommendations = [item for index, item in enumerate(df11) if index % 2 != 0]
recommendations
```

```python
#remove duplicates for earnings
df22 = [item for index, item in enumerate(df2) if index % 3 != 0]
earnings = [item for index, item in enumerate(df22) if index % 2 != 0]
earnings
```

```python
#convert string to float
recom=[float(i) for i in recommendations]
earnings=[float(i) for i in earnings]
```
